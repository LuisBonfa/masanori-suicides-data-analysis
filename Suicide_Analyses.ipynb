```python
import pandas as pd
from pandas import Series, DataFrame
df= pd.DataFrame(pd.read_csv("datasets/who_suicide_statistics.csv"))

    
    
```

```python
df
```

```python
brazil = df.loc[df['country'] == 'Brazil']
```

```python
brazil
```

```python
females = brazil.loc[brazil['sex'] == 'female']
males = brazil.loc[brazil['sex'] == 'male']
yearsTotalFemales = females['year'].values

countFemales = females["suicides_no"].sum()
countMales = males["suicides_no"].sum()

females,males
```

```python
from matplotlib.ticker import FuncFormatter
import matplotlib.pyplot as plt
import numpy as np

years = ['5-14 years',
        '15-24 years',
        '25-34 years',
        '35-54 years',
        '55-74 years',
        '75+ years']

data = []
total = 0    

for year in years:
    sum = females.loc[females['age'] == year]
    data.append(sum["suicides_no"].values.sum())
    total += sum["suicides_no"].values.sum()


fig, ax = plt.subplots(figsize=(15, 15), subplot_kw=dict(aspect="equal"))


def func(pct, allvals):
    absolute = int(pct/100.*np.sum(allvals))
    return "{:.1f}%\n({:d} Mulheres)".format(pct, absolute)


wedges, texts, autotexts = ax.pie(data, autopct=lambda pct: func(pct, data),
                                  textprops=dict(color="w"))

ax.legend(wedges, years,
          title="Idade",
          loc="center left",
          bbox_to_anchor=(1, 0, 0.5, 1))

plt.setp(autotexts, size=8, weight="bold")
    
ax.set_title("Suicídios Cometidos por Mulheres no Brasil desde 1979\n Total de: " + str(int(total)) )

plt.show()
```

```python
countries = ["France", "Argentina", "Brazil", "United States of America", "Germany", "Ireland", "Norway", "Mexico", 
            "Paraguay", "Venezuela (Bolivarian Republic of)", "Chile"]
sums = []
for country in countries:
    sumAux = df.loc[df['country'] == country]
    sumAux = sumAux["suicides_no"].dropna().values.sum()
    sums.append(sumAux)
   
f, ax = plt.subplots(figsize=(30,8))

plt.bar(countries, sums, label="Países")
plt.title("Comparação de Suicídios no Brasil com outros países")
ax.legend(fontsize=22)
```
